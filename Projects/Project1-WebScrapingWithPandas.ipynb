# 2 Read .csv from Multiple URLs with Pandas

##### https://www.football-data.co.uk/mmz4281/2425/E0.csv

##### Link: root + season + league

```python
# link structure
"https://www.football-data.co.uk/mmz4281/" + "2122" + "/" + "E0" + ".csv"
```

```python
# create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

## 2.1 Multiple Leagues

```python
# creating list of leagues
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']

frames = []
# looping through leagues, read multiple csv and append it into a list
for league in leagues:
    df = pd.read_csv( root + '2425' + '/' + league + '.csv')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st, 2nd and 3rd element
frames[4]
```

## 2.2 Multiple Seasons 

```python
# eg.
for season in range(15, 21):
    print(str(season) + str(season+1))
```

```python
# creating list of leagues
leagues = ['E0', 'E2', 'E3']
frames = []

# loop through range of seasons and insert new column
for league in leagues:
    for season in range(15, 21):
        df = pd.read_csv(f"{root}{season}{season+1}/{league}.csv")
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# total frames (6 seasons x 3 leagues)
len(frames)
```

```python
# show 1st, and last element (new season column was created)
frames[17]
```

# 3 Organize All The Data in a Dictionary

```python
# create a dictionary with original name of legue as key
dict_countries = {
                'Spanish La Liga':'SP1', 
                'Spanish Segunda Division':'SP2',
                'German Bundesliga':'D1', 
                'English Premier League':'E0', 
                'English League 1':'E2', 
                'English League 2':'E3',
            }
```

```python
# get a dictionary element
dict_countries['Spanish La Liga']
```

```python
# This will hold the combined historical data for each league
dict_historical_data = {}

# Loop over each league name in the dictionary
for league in dict_countries:
    frames = [] # List to store DataFrames for each season of the current league

    # Loop over the seasons from 2015 to 2020
    for season in range(15, 21):
        df= pd.read_csv(f"{root}{season}{season+1}/{dict_countries[league]}.csv")
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```

```python
# show dataframe inside dict_historical_data
dict_historical_data['English League 1']
```
