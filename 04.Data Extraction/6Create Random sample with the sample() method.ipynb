```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace=True)
# select column
df = df[['long_name', 'age','dob', 'height_cm',	'weight_kg', 'nationality', 'club']]
```

## 1 Sample()

```python
# extract 10 random elements from the "nationality" column
df['nationality'].sample(10)
```

```python
# extract a random 20% sample of the dataframe
df.sample(frac=0.2)
```

```python
# upsample: increase the sampling rate -> frac>1 (Note: replace patameter has to be True for frac parameter > 1)
df.sample(frac=2, replace=True)
```
