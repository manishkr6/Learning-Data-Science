```python
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

# 1 Rename Column

```python
# rename column and overwrite valuse of the dataframe
df_exams.rename(columns={'gender': 'Gender'})
```

```python
# rename multiple columns and update the dataframe with inplace argument
df_exams.rename(columns={'math score': 'MS',
                        'reading score': 'RS',
                        'writing score': 'WS'}, inplace=True)
```

```python
# show dataframe
df_exams
```

# 2 Rename Index

```python
# rename index 0, 1 and 2 and update the dataframe
df_exams.rename(index={0: 'A', 1: 'B', 2: 'C'}, inplace=True)
```
